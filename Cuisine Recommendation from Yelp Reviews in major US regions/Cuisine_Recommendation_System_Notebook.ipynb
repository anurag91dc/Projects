```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
 
# function to print sentiments
# of the sentence.
def sentiment_scores(sorted_reviews):
 
    # Create a SentimentIntensityAnalyzer object.
    sid_obj = SentimentIntensityAnalyzer()
 
    # polarity_scores method of SentimentIntensityAnalyzer
    # object gives a sentiment dictionary.
    # which contains pos, neg, neu, and compound scores.
    sentiment_score_array = []
    print(type(sentiment_score_array))
    for i in sorted_reviews:
      sentiment_dict = sid_obj.polarity_scores(i)
      sentiment_score_array.append(sentiment_dict['compound'])
    print(sentiment_score_array)
    sorted_df['sentiment_score'] = sentiment_score_array

sentiment_scores(sorted_reviews)
```

```python
#@title Default title text
df = sorted_df
def replace_attr (attribute_list, current_attribute):
  for i in attribute_list:
    df['reviews'] = df['reviews'].str.replace(i,current_attribute)

replace_attr(Italian,'Italian')
replace_attr(American,'American')
replace_attr(Asian,'Asian')
replace_attr(Texmex,'Texmex')
replace_attr(Mediterranean,'Mediterranean')
replace_attr(French,'French')
replace_attr(tasty,'tasty')
replace_attr(price,'price')
replace_attr(bad,'bad')
replace_attr(portion,'portion')
replace_attr(seating,'seating')
replace_attr(beverages,'beverages')
replace_attr(drinks,'drinks')

df['reviews'].iloc[3]
```

```python
import math
def dot(v,w):
    """ v_1*w_1 + ... v_n*w_n"""
    return sum(v_i *w_i
              for v_i,w_i in zip(v,w))
def cosine_similarity(v, w):
    return 0 if (math.sqrt(dot(v, v) * dot(w, w)) == 0) else dot(v, w) / math.sqrt(dot(v, v) * dot(w, w))
```

```python
df['reviews'] = df['reviews'].map(str).apply(lambda x: x.lower())

def attrbute_counts (attribute_list, current_attribute):
  df[current_attribute] = [0 for i in range(0, len(df['reviews']))]
  for attribute in attribute_list:
    df[current_attribute] += df['reviews'].apply(lambda x: 1 if (len(re.findall(attribute, x)) > 0) else 0)

attrbute_counts(['texmex'], 'texmex')
attrbute_counts(['price'], 'price')
attrbute_counts(['portion'], 'portion')

df
```

We use 3 attributes to provide a reccommendation to the Client

```python
df_similarity = df[['texmex','price','portion']]
customer_vectors = df_similarity.to_numpy()
review_similarities = [cosine_similarity([1,1,1], vector) for vector in customer_vectors]

df['similarity_score'] = review_similarities
df
```
